# Barcode Sharing Troubleshooting
This notebook plots progeny barcode frequencies. It checks whether viral barcodes are found in `WT` data, `syn` data, or both. In addition, it looks for barcodes in the original WT and syn rescued virus barcode data, as well.

Data for **supernatant and second infection** are from `scProgenyProduction_trial1`.

## Notebook setup

Import Python modules:

```python
import gzip
import itertools
import random

from IPython.display import display

from dms_variants.constants import CBPALETTE

import editdistance

import numpy as np

import pandas as pd

import plotnine as p9
from mizani.formatters import percent_format
```

Variables are hardcoded by `snakeamke` in first cell. Specify sample variables.

```python
viral_bc_in_progeny_csv = 'results/viral_progeny/scProgenyProduction_trial1_bcTroubleShooting_viral_bc_in_progeny.csv.gz'
expt = 'bc_sharing_troubleshooting_scProgenyProduction_trial2'
```

## Process data

Load viral barcode counts and calculate the frequency of each barcode within a sequencing sample.

```python
viral_bc = pd.read_csv(gzip.open(viral_bc_in_progeny_csv))
```

```python
viral_bc = pd.merge(viral_bc,
                    (viral_bc
                     .groupby(['source',
                               'tag',
                               'gene',
                               'replicate'])
                     .sum()
                     .reset_index()),
                    on=['source', 'tag', 'gene', 'replicate'],
                    suffixes=('_barcode', '_total'))
viral_bc['freq'] = (viral_bc['count_barcode'] /
                    viral_bc['count_total'])
display(viral_bc)
```

Check whether each barcode is found in wt and syn rescue virus data:

```python
rescue_barcodes = ({
                    'fluHA_wt': (set(viral_bc
                                       .query('(source == "rescue") '
                                              'and (tag == "wt") '
                                              'and (gene == "fluHA")')
                                       ['barcode']
                                       .unique())
                                      ),
                    'fluHA_syn': (set(viral_bc
                                       .query('(source == "rescue") '
                                              'and (tag == "syn") '
                                              'and (gene == "fluHA")')
                                       ['barcode']
                                       .unique())
                                      ),
                    'fluNA_wt': (set(viral_bc
                                       .query('(source == "rescue") '
                                              'and (tag == "wt") '
                                              'and (gene == "fluNA")')
                                       ['barcode']
                                       .unique())
                                      ),
                    'fluNA_syn': (set(viral_bc
                                       .query('(source == "rescue") '
                                              'and (tag == "syn") '
                                              'and (gene == "fluNA")')
                                       ['barcode']
                                       .unique()))
})

def check_rescue(row):
    gene = row['gene']
    barcode = row['barcode']
    tags = list()
    for tag in ['wt', 'syn']:
        if barcode in rescue_barcodes[f'{gene}_{tag}']:
            tags.append(tag)
        else:
            pass
    if not tags:
        return 'neither'
    else:
        tag_str = ' and '.join(sorted(set(tags)))
        return(tag_str)

viral_bc['rescue_tags'] = viral_bc.apply(check_rescue, axis=1)

display(viral_bc)
```

```python
tag_freqs = (viral_bc
             .pivot_table(index=['gene',
                                 'tag',
                                 'barcode',
                                 'tags_present'],
                          columns=['source'],
                          values='freq')
             .fillna(0)
             .reset_index())

display(tag_freqs)

fig = (p9.ggplot(tag_freqs,
                 p9.aes(x='plasmid',
                        y='rescue',
                        color='tags_present')) +
       p9.ggtitle(f'{expt}\n'
                  'barcode counts in plasmid '
                  'and resced virus') +
       p9.geom_point(alpha=0.4) +
       p9.facet_grid('gene~tag') +
       p9.theme(figure_size=(4, 4),
                plot_title=p9.element_text(size=10),
                axis_title=p9.element_text(size=10),
                legend_title=p9.element_text(size=9),
                legend_title_align='center') +
       p9.scale_x_log10() +
       p9.scale_y_log10() +
       p9.scale_color_manual(CBPALETTE[0:]))
fig
```

Plot correlation of barcodes in supernatant and in second infection. Color by presence in each (or both) tags.

```python
fig = (p9.ggplot(tag_freqs,
                 p9.aes(x='supernatant',
                        y='second_infection',
                        color='tags_present')) +
       p9.ggtitle(f'{expt}\n'
                  'barcode counts in supernatant '
                  'and second infection') +
       p9.geom_point(alpha=0.4) +
       p9.facet_grid('gene~tag') +
       p9.theme(figure_size=(4, 4),
                plot_title=p9.element_text(size=10),
                axis_title=p9.element_text(size=10),
                legend_title=p9.element_text(size=9),
                legend_title_align='center') +
       p9.scale_x_log10() +
       p9.scale_y_log10() +
       p9.scale_color_manual(CBPALETTE[0:]))
fig
```
