```python
viral_tag_by_cell_csv = (snakemake
                         .input
                         .viral_tag_by_cell_csv)
viral_bc_by_cell_corrected_csv = (snakemake
                                  .input
                                  .viral_bc_by_cell_corrected_csv)
viral_genes_by_cell_csv = 'results/viral_fastq10x/scProgenyProduction_trial3_viral_genes_by_cell.csv.gz'
viral_bc_by_cell_csv = 'results/viral_fastq10x/scProgenyProduction_trial3_viral_bc_by_cell.csv.gz'
plot = snakemake.output.plot
expt = snakemake.wildcards.expt
barcoded_viral_genes = snakemake.params.barcoded_viral_genes
fdr = snakemake.params.fdr
```

Set base plot style:

```python
p9.theme_set(p9.theme_classic())
```

## Organize data

Read the viral barcode UMI counts data into a pandas dataframe:

```python
viral_bc_counts = pd.read_csv(viral_bc_by_cell_corrected_csv)
viral_bc_counts = (viral_bc_counts
                   .rename(columns={'count': 'viral_bc_UMIs'}))
assert set(viral_bc_counts['gene']) == set(barcoded_viral_genes), \
       "Barcoded genes in barcode counts do not match expectation."
display(viral_bc_counts)
```

Read the total number of UMIs per cell into a pandas dataframe. Only keep relevant columns.

```python
all_cells = pd.read_csv(viral_tag_by_cell_csv)
all_cells = all_cells[['cell_barcode',
                       'infected',
                       'infecting_viral_tag',
                       'total_UMIs',
                       'viral_UMIs',
                       'cellular_UMIs',
                       'frac_viral_UMIs']]
display(all_cells)
```

Sanity check that `total_UMIs` is equal to `viral_UMIs + cellular_UMIs`:

```python
assert bool((all_cells['total_UMIs'] ==
             all_cells['viral_UMIs'] +
             all_cells['cellular_UMIs'])
            .all()), "UMI counts do not add up"
```

Read the gene presence/absence data

```python
viral_genes_by_cell = pd.read_csv(viral_genes_by_cell_csv)
display(viral_genes_by_cell)
```

Merge dataframes:

```python
viral_bc_frac = pd.merge(
    left=pd.concat([all_cells.assign(gene=gene)
                    for gene in barcoded_viral_genes]),
    right=viral_bc_counts,
    how='outer',
    on=['cell_barcode', 'gene'],
    validate='one_to_many')
assert (viral_bc_frac['cell_barcode'].unique() ==
        all_cells['cell_barcode'].unique()).all(), \
       "Cell barcodes in merged dataframe don't " \
       "match barcodes in source data."
assert (viral_bc_frac['viral_barcode'].nunique() ==
        viral_bc_counts['viral_barcode'].nunique()), \
       "Number of viral barcodes in merged dataframe don't " \
       "match number of barcodes in source data."

# Make `infecting_viral_tag` column ordered category
viral_bc_frac['infecting_viral_tag'] = (viral_bc_frac['infecting_viral_tag']
                                        .astype('category')
                                        .cat
                                        .reorder_categories(['none',
                                                             'wt',
                                                             'syn',
                                                             'both']))

viral_bc_frac = pd.merge(
    left=viral_bc_frac,
    right=viral_genes_by_cell,
    on=['cell_barcode', 'gene'],
    how='left',
    validate='many_to_one'
)

display(viral_bc_frac)
```

Calculate **each barcode's** fraction of all UMIs per cell:

```python
viral_bc_frac = (
    viral_bc_frac
    .assign(viral_bc_UMIs=lambda x: (x['viral_bc_UMIs']
                                     .fillna(0)
                                     .astype(int, errors='raise'))))


viral_bc_frac['frac_viral_bc_UMIs'] = (
    viral_bc_frac['viral_bc_UMIs'] /
    viral_bc_frac['total_UMIs'])

display(viral_bc_frac)
```

**For each cell, for each gene,** sum the number of viral barcode UMIs.

```python
viral_bc_frac = (
    viral_bc_frac
    .assign(sum_UMIs_with_viral_bc_for_cell_and_gene=(
        lambda x: ((x
                    .groupby(['cell_barcode', 'gene'])
                    ['viral_bc_UMIs'])
                   .transform('sum')))))

display(viral_bc_frac)
```

Calculate fraction of all UMIs and fraction of viral UMIs that have barcode for each cell-gene.

```python
viral_bc_frac['frac_total_UMIs_with_viral_bc_for_cell_and_gene'] = (
    viral_bc_frac['sum_UMIs_with_viral_bc_for_cell_and_gene'] /
    viral_bc_frac['total_UMIs'])
display(viral_bc_frac.groupby(['cell_barcode', 'gene']).head())
```

Extract number of infected cells (this is useful for visualization and getting a sense of where filters lie compared to infection calls)

```python
n_infected_cells = (viral_bc_frac.query('infected == "infected"')
                    ['cell_barcode']
                    .nunique())
print(f"There are {n_infected_cells} infected cells in dataset.")
```

## Filtering
Now, I will filter out viral barcodes in each cell. A valid viral barcode meets the following criteria:
1. The cell expresses the barcoded gene
2. There are at least 2 UMIs with the viral barcode in that cell-gene
3. The viral barcode is present at a frequency greater than some cutoff percentile observed in uninfected cells.

```python
filter_query = str()
```

### Gene expression
Plot total barcode UMIs per cell annotated with gene expression:

```python
frac_total_rank = (viral_bc_frac[['cell_barcode',
                                  'infected',
                                  'gene',
                                  'gene_present',
                                  'frac_total_UMIs_with_viral_bc_for_cell_and_gene',]]
                   .copy()
                   .drop_duplicates())
frac_total_rank['rank'] = (frac_total_rank
                           .groupby('gene')
                           ['frac_total_UMIs_with_viral_bc_for_cell_and_gene']
                           .rank(ascending=False, method='first'))

fig = (p9.ggplot(frac_total_rank,
                 p9.aes(x='rank',
                        y='frac_total_UMIs_with_viral_bc_for_cell_and_gene',
                        color='gene_present')) +
       p9.geom_point(alpha=0.5) +
       p9.geom_vline(xintercept=n_infected_cells, linetype='dashed', color='red') +
       p9.facet_grid('gene~') +
       p9.scale_y_log10() +
       p9.ggtitle('total fraction of UMIs with viral bc\n'
                  f'{expt}') +
       p9.labs(x='cell rank') +
       p9.theme(figure_size=(4, 4),
                plot_title=p9.element_text(size=9),
                axis_title=p9.element_text(size=9),
                legend_title=p9.element_text(size=9),
                legend_title_align='center') +
       p9.scale_color_manual(CBPALETTE[0:]))
display(fig)
```

Add gene expression to filter query:

```python
filter_query = (
    filter_query +
    '(gene_present == True) and '
)
print("Filter query is: \n"
      f"{filter_query}")
```

### Absolute abundance
Label and plot viral barcodes with fewer than 2 UMIs in a given cell:

```python
min_UMI = 2
viral_bc_frac['above_UMI_limit'] = viral_bc_frac['viral_bc_UMIs'] >= min_UMI

UMI_limit_rank = (viral_bc_frac[['cell_barcode',
                                  'infected',
                                  'gene',
                                  'above_UMI_limit',
                                  'viral_bc_UMIs',]]
                   .copy()
                   .drop_duplicates())
UMI_limit_rank['rank'] = (UMI_limit_rank
                           .groupby('gene')
                           ['viral_bc_UMIs']
                           .rank(ascending=False, method='first'))

fig = (p9.ggplot(UMI_limit_rank,
                 p9.aes(x='rank',
                        y='viral_bc_UMIs',
                        color='above_UMI_limit')) +
       p9.geom_point(alpha=0.5) +
       p9.geom_hline(yintercept=min_UMI, linetype='dashed', color=CBPALETTE[2]) +
       p9.geom_vline(xintercept=n_infected_cells, linetype='dashed', color='red') +
       p9.facet_grid('gene~') +
       p9.scale_y_log10() +
       p9.ggtitle(f'{expt}') +
       p9.labs(x='viral bc rank') +
       p9.theme(figure_size=(4, 4),
                plot_title=p9.element_text(size=9),
                axis_title=p9.element_text(size=9),
                legend_title=p9.element_text(size=9),
                legend_title_align='center') +
       p9.scale_color_manual(CBPALETTE[0:]))
display(fig)
```

Add absolute abundance to filter query:

```python
filter_query = (
    filter_query +
    '(above_UMI_limit == True) and '
)
print("Filter query is: \n"
      f"{filter_query}")
```

### Relative abundance
Find the 99th percentile of individual viral barcode frequency in uninfected cells. Label and filter out viral barcodes that are less than this frequency.

```python
limit_percentile = 0.99

uninfected_background_frac = {}
for gene in barcoded_viral_genes:
    limit = (viral_bc_frac
             .query('(infecting_viral_tag == "none") '
                    f'and (gene == "{gene}")')
             ['frac_viral_bc_UMIs']
             .quantile(limit_percentile))
    uninfected_background_frac[f'{gene}'] = limit
    print(f'The limit for a viral barcode on {gene} is {limit:.5f}.')
    
# Convert to DataFrame for plotting
uninfected_background_frac = (pd.DataFrame.from_dict(uninfected_background_frac,
                                                orient='index',
                                                columns=['uninf_background'])
                         .reset_index()
                         .rename(columns={'index': 'gene'}))

fig = (p9.ggplot((viral_bc_frac
                  .query('infecting_viral_tag == "none"')),
                 p9.aes(x='frac_viral_bc_UMIs')) +
       p9.geom_histogram(bins=50, fill="#3a3a3a") +
       p9.geom_vline(uninfected_background_frac,
                     p9.aes(xintercept='uninf_background'),
                     linetype='dashed',
                     color='#3A3B3C',
                     size=0.5) +
       p9.facet_grid('infecting_viral_tag~gene') +
       p9.ggtitle('viral barcode fraction in uninfected cells\n'
                  f'{expt}') +
       p9.labs(x='each viral barcode\'s fraction of total UMIs',) +
       p9.theme(figure_size=(5, 1),
                plot_title=p9.element_text(size=10),
                axis_title=p9.element_text(size=10),
                axis_text=p9.element_text(rotation=45),
                legend_title=p9.element_text(size=9),
                legend_title_align='center'))
display(fig)
```

```python
viral_bc_frac = pd.merge(left=viral_bc_frac,
                         right=uninfected_background_frac,
                         on='gene',
                         how='outer',
                         validate='many_to_one')
viral_bc_frac['above_uninf_background'] = (viral_bc_frac['frac_viral_bc_UMIs'] >
                                           viral_bc_frac['uninf_background'])
```

```python
frac_rank = (viral_bc_frac[['cell_barcode',
                            'infected',
                            'gene',
                            'above_uninf_background',
                            'frac_viral_bc_UMIs',]]
                   .copy()
                   .drop_duplicates())
frac_rank['rank'] = (frac_rank
                     .groupby('gene')
                     ['frac_viral_bc_UMIs']
                     .rank(ascending=False, method='first'))

fig = (p9.ggplot(frac_rank,
                 p9.aes(x='rank',
                        y='frac_viral_bc_UMIs',
                        color='above_uninf_background')) +
       p9.geom_point(alpha=0.5) +
       p9.geom_hline(uninfected_background_frac,
                     p9.aes(yintercept='uninf_background'),
                     linetype='dashed',
                     color=CBPALETTE[2]) +
       p9.geom_vline(xintercept=n_infected_cells, linetype='dashed', color='red') +
       p9.facet_grid('gene~') +
       p9.scale_y_log10() +
       p9.ggtitle(f'{expt}') +
       p9.labs(x='viral bc rank') +
       p9.theme(figure_size=(4, 4),
                plot_title=p9.element_text(size=9),
                axis_title=p9.element_text(size=9),
                legend_title=p9.element_text(size=9),
                legend_title_align='center') +
       p9.scale_color_manual(CBPALETTE[0:]))
display(fig)
```

Same plot, colored by infection status to show that threshold removes frequencies compatible with contamination rate in uninfected cells

```python
fig = (p9.ggplot(frac_rank,
                 p9.aes(x='rank',
                        y='frac_viral_bc_UMIs',
                        color='infected')) +
       p9.geom_point(alpha=0.5) +
       p9.geom_hline(uninfected_background_frac,
                     p9.aes(yintercept='uninf_background'),
                     linetype='dashed',
                     color=CBPALETTE[2]) +
       p9.geom_vline(xintercept=n_infected_cells, linetype='dashed', color='red') +
       p9.facet_grid('gene~') +
       p9.scale_y_log10() +
       p9.ggtitle(f'{expt}') +
       p9.labs(x='viral bc rank') +
       p9.theme(figure_size=(4, 4),
                plot_title=p9.element_text(size=9),
                axis_title=p9.element_text(size=9),
                legend_title=p9.element_text(size=9),
                legend_title_align='center') +
       p9.scale_color_manual(CBPALETTE[0:]))
display(fig)
```

Add background frequency to filter query:

```python
filter_query = (
    filter_query +
    '(above_uninf_background == True)'
)
print("Filter query is: \n"
      f"{filter_query}")
```

## Assign filter status to each viral barcode

```python
viral_bc_frac.loc[viral_bc_frac.eval(filter_query), 'confident_viral_bc'] = True
viral_bc_frac['confident_viral_bc'] = viral_bc_frac['confident_viral_bc'].fillna(False)
display(viral_bc_frac)
```

```python
fig = (p9.ggplot(viral_bc_frac,
                 p9.aes(x='frac_viral_UMIs',
                        y='frac_viral_bc_UMIs',
                        color='confident_viral_bc')) +
       p9.geom_point(alpha=0.2) +
       p9.facet_grid('gene~') +
       p9.scale_x_log10() + p9.scale_y_log10() +
       p9.ggtitle('viral burden and viral bc frequency\n'
                  f'{expt}') +
       p9.labs() +
       p9.theme(figure_size=(2, 4),
                plot_title=p9.element_text(size=9),
                axis_title=p9.element_text(size=9),
                legend_title=p9.element_text(size=9),
                legend_title_align='center') +
       p9.scale_color_manual(CBPALETTE[0:]))
display(fig)
```

## Calculate rate of missed and spurious barcodes
Want to know how many cells called as uninfected have >= 1 viral barcodes called as statistically significant.  
Likewise, want to know how many cells called as infected have 0 viral barcodes called as statistically significant.

I will check whether each cell-gene has any viral barcodes that are called as significant (`confident_viral_bc`). This call will be put in a column called `confident_viral_bc_present_in_cell_gene`.

```python
confident_viral_bc_counts = (
    viral_bc_frac
    .groupby(['cell_barcode',
              'infected',
              'gene'])
    ['confident_viral_bc']
    .any()
    .dropna()
    .reset_index()
    .rename(columns={
        'confident_viral_bc': 'confident_viral_bc_present_in_cell_gene'}))

# Check that length of significant_viral_bc_counts is 2X length of all cells
assert len(confident_viral_bc_counts) == 2 * len(all_cells), \
    "Number of cells input does not match number of cell-gene " \
    "combinations generated"


confident_viral_bc_counts
```

```python
output_fig = (p9.ggplot(
        confident_viral_bc_counts,
        p9.aes(x='infected',
               fill='confident_viral_bc_present_in_cell_gene')) +
              p9.geom_bar(stat='count', position='dodge') +
              p9.ggtitle('confident viral bc identified in cell-gene\n'
                         f'{expt}') +
              p9.ylab('n_cells') +
              p9.facet_grid('~gene') +
              p9.theme(figure_size=(4, 2),
                       plot_title=p9.element_text(size=10),
                       axis_title=p9.element_text(size=10),
                       axis_text=p9.element_text(rotation=45),
                       legend_title=p9.element_text(size=9),
                       legend_title_align='center') +
              p9.scale_fill_manual(CBPALETTE[0:]))
display(output_fig)
```

```python
print(f"Saving plots to {plot}")
p9.ggsave(plot=output_fig, filename=plot, verbose=False)

print(f"Saving filtered barcodes to {viral_bc_by_cell_csv}")

# Export all barcodes, and annotate whether uninfected null hypothesis
# is rejected. Could imagine filtering on other criteria in future.
viral_bc_frac.to_csv(viral_bc_by_cell_csv,
                    columns=['cell_barcode',
                              'gene',
                              'viral_barcode',
                              'viral_bc_UMIs',
                              'frac_viral_bc_UMIs',
                              'confident_viral_bc'],
                     index=False)
```
